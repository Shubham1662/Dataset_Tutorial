# **Loading PDF Document using PyPDF Loader**

```python
import os
import requests
url = "https://raw.githubusercontent.com/Shubham1662/Dataset_Tutorial/main/budget_speech.pdf"

# Download PDF
response = requests.get(url)
with open("budget_speech.pdf", "wb") as f:
    f.write(response.content)

# Load PDF
loader = PyPDFLoader("budget_speech.pdf")
documents = loader.load()

len(documents)
```

# **We can see the document is loaded and total no. of pages in the document.**

```python
print(documents)
```

# **We will divide the whole document into smaller chunks with Overlap.**

```python
## Divide the docs into chunks
def chunk_data(docs,chunk_size=800,chunk_overlap=80):
    text_splitter=RecursiveCharacterTextSplitter(chunk_size=chunk_size,chunk_overlap=chunk_overlap)
    doc=text_splitter.split_documents(docs)
    return doc
```

# We can see now that total number of document is now increased to 169 because chunk size is smaller than content in 1

```python
documents=chunk_data(docs=documents)
len(documents)
```

```python
for i in documents:

  print("****************************************")
  print("New Chunk starting")
  print(" ")
  print(i)
  print("****************************************")
  print("Chunk Ended")
  print(" ")
```

```python
index_name="ragtutor"
```

```python
from langchain_pinecone import PineconeVectorStore
```

```python
index = PineconeVectorStore.from_documents(documents=documents, embedding=embeddings, index_name=index_name)
```

```python
## Cosine Similarity Retreive Results from VectorDB
def retrieve_query(query,k=2):
    matching_results=index.similarity_search(query,k=k)
    return matching_results
```

### **# In artificial intelligence (AI) and machine learning, temperature is a parameter for adjusting the output of large language models (LLMs). Temperature controls the randomness of text that is generated by LLMs during inference.**

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.9)
```

```python
from langchain_classic.chains.question_answering import load_qa_chain
chain = load_qa_chain(llm, chain_type="stuff")
```

```python
## Search answers from VectorDB
def retrieve_answers(query):
    doc_search=retrieve_query(query)
    #print(doc_search)
    response=chain.run(input_documents=doc_search,question=query)
    return response
```

```python
our_query = "I run a pharma business. What is in it for me.  Answer in less than 35 words"
answer = retrieve_answers(our_query)
print(answer)
```
